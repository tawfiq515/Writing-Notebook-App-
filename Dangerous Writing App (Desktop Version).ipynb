```python
import tkinter as tk
from tkinter import filedialog, messagebox

class DangerousWritingApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Dangerous Writing App")
        self.root.geometry("600x400")

        self.timer = None
        self.time_limit = 5000

        self.menubar = tk.Menu(root)
        self.filemenu = tk.Menu(self.menubar, tearoff=0)
        self.filemenu.add_command(label="Save", command=self.save_file)
        self.filemenu.add_command(label="Save As...", command=self.save_file_as)
        self.filemenu.add_separator()
        self.filemenu.add_command(label="Exit", command=root.quit)
        self.menubar.add_cascade(label="File", menu=self.filemenu)
        root.config(menu=self.menubar)

        self.text_area = tk.Text(root, font=("Arial", 14), wrap="word")
        self.text_area.pack(expand=True, fill="both")
        self.text_area.bind("<Key>", self.reset_timer)

        self.current_file = None
        self.start_timer()

    def start_timer(self):
        self.timer = self.root.after(self.time_limit, self.clear_text)

    def reset_timer(self, event=None):
        if self.timer:
            self.root.after_cancel(self.timer)
        self.start_timer()

    def clear_text(self):
        if messagebox.askyesno("Clear Text", "You've stopped writing! Clear all text?"):
            self.text_area.delete("1.0", tk.END)

    def save_file(self):
        if self.current_file:
            try:
                with open(self.current_file, 'w') as f:
                    f.write(self.text_area.get("1.0", tk.END))
            except:
                self.save_file_as()
        else:
            self.save_file_as()

    def save_file_as(self):
        file = filedialog.asksaveasfile(defaultextension=".txt",
                                       filetypes=[("Text Files", "*.txt"),
                                                  ("All Files", "*.*")])
        if file:
            self.current_file = file.name
            file.write(self.text_area.get("1.0", tk.END))
            file.close()

if __name__ == "__main__":
    root = tk.Tk()
    app = DangerousWritingApp(root)
    root.mainloop()
```
